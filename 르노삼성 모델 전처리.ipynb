```python
df['trim']= pd.Series()
```

```python
for i in range(len(df['info'])):
    if "PE" in df['info'][i]:
        df["trim"][i] = "1"
    elif "SE 플러스" in df['info'][i]:
        df["trim"][i] = "1"
    elif "SE 블랙" in df['info'][i]:
        df["trim"][i] = "1"
    elif "SE 플레져" in df['info'][i]:
        df["trim"][i] = "1"
    elif "SE" in df['info'][i]:
        df["trim"][i] = "1"
    elif "XE" in df['info'][i]:
        df["trim"][i] = "1"
    elif "LE 플러스" in df['info'][i]:
        df["trim"][i] = "2"
    elif "LE 익스클루시브" in df['info'][i]:
        df["trim"][i] = "2"
    elif "LE" in df['info'][i]:
        df["trim"][i] = "2"
    elif "RE 시그니처" in df['info'][i]:
        df["trim"][i] = "3"
    elif "RE" in df['info'][i]:
        df["trim"][i] = "3"
    elif "Premiere" in df['info'][i]:
        df["trim"][i] = "4"
    elif "인텐스" in df['info'][i]:
        df["trim"][i] = "2"
    elif "프레스티지" in df['info'][i]:
        df["trim"][i] = "4"
    elif "클래식" in df['info'][i]:
        df["trim"][i] = "1"
    elif "씨티" in df['info'][i]:
        df["trim"][i] = "1"
    elif "D 프리미엄" in df['info'][i]:
        df["trim"][i] = "1"
        
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['info'], axis=1,inplace=True)
df.dropna(axis=0,inplace=True)
```

```python
df['type'] = df.apply(lambda x: x['type'].replace('(일반인 구입)',''), axis = 1)
df['type'] = df.apply(lambda x: x['type'].replace('가솔린+전기','하이브리드'), axis = 1)
df['type'] = df.apply(lambda x: x['type'].strip(), axis = 1)
```
