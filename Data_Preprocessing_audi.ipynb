```python
df = pd.read_csv('data/아우디-Copy1.csv')
df
```

```python
# 쓸모없는 컬럼 제거
df.drop(['Unnamed: 0'],axis=1, inplace = True)
df
```

```python
# info 행의 사이드 콰트로 문자열 제거
df['info'] = df.apply(lambda x: x['info'].strip('콰트로'), axis = 1)
df
```

```python
# info 행의 중간 콰트로 문자열 제거
df['info'] = df.apply(lambda x: x['info'].replace('콰트로',''), axis = 1)
df
```

```python
df['info'].unique()
```

```python
# info열의 프리미엄 스포츠 여부를 upgrade 열에 추가
df['upgrade'] = ""

for i in range(0, len(df)):
    if df['info'].str.contains('엔트리')[i]:
        df['upgrade'][i] = 1
    elif df['info'].str.contains('스포트라인')[i]:
        df['upgrade'][i] = 1
    
    elif df['info'].str.contains('다이나믹')[i]:
        df['upgrade'][i] = 1
        
    elif df['info'].str.contains('리미티드')[i]:
        df['upgrade'][i] = 2
        
    elif df['info'].str.contains('프레스티지')[i]:
        df['upgrade'][i] = 2
        
    elif df['info'].str.contains('프리미엄')[i]:
        df['upgrade'][i] = 2
        
    elif df['info'].str.contains('S-LINE')[i]:
        df['upgrade'][i] = 3
        
    elif df['info'].str.contains('S-line')[i]:
        df['upgrade'][i] = 3
        
    elif df['info'].str.contains('S-Line')[i]:
        df['upgrade'][i] = 3
        
    elif df['info'].str.contains('쿠페')[i]:
        df['upgrade'][i] = 4
        
    elif df['info'].str.contains('카브리올레')[i]:
        df['upgrade'][i] = 4
        
    elif df['info'].str.contains('스포츠')[i]:
        df['upgrade'][i] = 4
        
    elif df['info'].str.contains('스파이더')[i]:
        df['upgrade'][i] = 4
        
    else :
        df['upgrade'][i] = 1
df
```

```python
df.drop(['info'], axis = 1, inplace = True)
```

```python
df['name'] = df.apply(lambda x: x['name'].replace('(4S)',''), axis = 1)
df['name'] = df.apply(lambda x: x['name'].replace('(8V)',''), axis = 1)
df['name'] = df.apply(lambda x: x['name'].replace('(8P)',''), axis = 1)
df['name'] = df.apply(lambda x: x['name'].replace('뉴',''), axis = 1)
df['name'] = df.apply(lambda x: x['name'].strip(), axis = 1)
df['name'].unique()
```

```python
df['type'].unique()
```

```python
df['type'] = df.apply(lambda x: x['type'].replace('가솔린+전기','하이브리드'), axis = 1)
df['type'] = df.apply(lambda x: x['type'].replace('디젤+전기','하이브리드'), axis = 1)
```

```python
df.columns = ['brand','name','type','km', 'year','accident','color','location','price','wd','trim']
df
```
