```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv("./data/제네시스.csv")
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head(30)
```

```python
# 데이터 확인 용도로 사용
df['info'].value_counts().to_csv("제네시스요약.csv")
```

```python
df['info']=df['info'].str.replace(pat='AWD', repl='')
df['info']=df['info'].str.replace(pat='2WD', repl='')
df['info']=df['info'].str.replace(pat='4WD', repl='')
df['info']=df['info'].str.replace(pat='가솔린', repl='')
df['info']=df['info'].str.replace(pat='디젤', repl='')
df['info']=df['info'].str.replace(pat='터보', repl='T')
df['info']=df['info'].str.replace(pat='2.2D', repl='2.2')
df['info']=df['info'].str.replace(pat=' T', repl='T')
df['info']=df['info'].str.replace(pat='GDi', repl='')
df['info']=df['info'].str.replace(pat='GDI', repl='')
df['info']=df['info'].str.replace(pat='-', repl='')
df['info']=df['info'].str.replace(pat='Turbo', repl='')
df['info']=df['info'].str.replace(pat='  ', repl='')
df['info']=df['info'].str.replace(pat='3.3프리미엄 럭셔리', repl='3.3 프리미엄 럭셔리')
df['info']=df['info'].str.replace(pat='3.3럭셔리', repl='3.3 럭셔리')
df['info']=df['info'].str.replace(pat='3.8럭셔리', repl='3.8 럭셔리')
df['info']=df['info'].str.replace(pat='3.8프레스티지', repl='3.8 프레스티지')
df['info']=df['info'].str.replace(pat='3.3프레스티지', repl='3.3 프레스티지')
df['info']=df['info'].str.strip()
```

```python
df['trim']= pd.Series()
```

```python
for i in range(len(df['info'])):
    if "3.3 프리미엄 럭셔리" in df['info'][i]:
        df["trim"][i] = "2"
    elif "3.3 프레스티지" in df['info'][i]:
        df["trim"][i] = "2"
    elif "3.3T 프리미엄 럭셔리" in df['info'][i]:
        df["trim"][i] = "2"
    elif "3.3T 스포츠" in df['info'][i]:
        df["trim"][i] = "3"
    elif "3.8 프리미엄 럭셔리" in df['info'][i]:
        df["trim"][i] = "2" 
    elif "3.8 럭셔리" in df['info'][i]:
        df["trim"][i] = "2" 
    elif "3.8 프레스티지" in df['info'][i]:
        df["trim"][i] = "3" 
    elif "5.0 리무진" in df['info'][i]:
        df["trim"][i] = "4" 
    elif "5.0 프레스티지" in df['info'][i]:
        df["trim"][i] = "4" 
    elif "럭셔리" in df['info'][i]:
        df["trim"][i] = "2" 
    elif "2.0T" in df['info'][i]:
        df["trim"][i] = "1"
    elif "2.0" in df['info'][i]:
        df["trim"][i] = "1"
    elif "2.2" in df['info'][i]:
        df["trim"][i] = "1"
    elif "2.5T" in df['info'][i]:
        df["trim"][i] = "1"
    elif "2.5" in df['info'][i]:
        df["trim"][i] = "1"
    elif "3.0" in df['info'][i]:
        df["trim"][i] = "1"
    elif "3.3T" in df['info'][i]:
        df["trim"][i] = "2"
    elif "3.3" in df['info'][i]:
        df["trim"][i] = "1"
    elif "3.5T" in df['info'][i]:
        df["trim"][i] = "1"
    elif "3.5" in df['info'][i]:
        df["trim"][i] = "1"
    elif "3.8" in df['info'][i]:
        df["trim"][i] = "1"
    elif "5.0" in df['info'][i]:
        df["trim"][i] = "1"
    elif "" in df['info'][i]:
        df["trim"][i] = "2"
        
df.head()
```

```python
df=df.groupby('info').filter(lambda x : len(x)>10)
```

```python
df.drop(['info'], axis=1,inplace=True)
```
