```python
# 색별로 분류
url = 'http://www.m-park.co.kr/buy/my_car_list.asp'
data ={'sSearch': 1,
    'sListOrder': 'DemoDay_D',
    'sColorCode': '006007',
    'sPageSize': 40,
    'gotoPage': 1}
res = requests.post(url,data=data)
bs_obj = BeautifulSoup(res.text,'html.parser')
```

```python
ul = bs_obj.find("ul", {"class":"theme_view on"})
prd_info = ul.findAll("div",{"class":"car-info-s"})
year= []
year_1 = []
km = []
fuel = []

for i in range(0,len(prd_info)) : 
    year.append(prd_info[i].text.split('\n',maxsplit = 4)[1])
    km.append(prd_info[i].text.split('\n',maxsplit = 4)[2])
    fuel.append(prd_info[i].text.split('\n',maxsplit = 4)[3])
year_1 = [year[i].replace('년식',"") for i in range(0,len(year))]


car_manu = []
car_model_1 = []
car_model_2 = []
car_price = []

prd_name = ul.findAll("div",{"class":"title-box"})
# 모델명 추출
car_model = [prd_name[i].find("h1").text for i in range(0,len(prd_name))]
car_model_2 = [prd_name[i].find("h2").text for i in range(0,len(prd_name))]
for i in range(0, len(car_model)) :
    car_manu.append(car_model[i].split("\xa0", maxsplit=1)[0])
    car_model_1.append(car_model[i].split("\xa0", maxsplit=1)[1])

# 가격
prd_price = ul.findAll("div",{"class":"price-info"})
# 모델명 추출
car_price = [prd_price[i].find("span",{"class":"point"}).text for i in range(0,len(prd_price))]

df = pd.DataFrame()
df['제조사'] = car_manu
df['모델명'] = car_model_1
df['옵션명'] = car_model_2
df['연식'] = year_1
df['주행거리'] = km
df['연료'] = fuel
df['색상'] = ['검정' for i in range(0,len(df))]
df['가격'] = car_price
df
```

```python
key = list(colorcode.keys())
value = list(colorcode.values())
```

```python
# 크롤링 색상별 함수화
def bring_obj(value, pagenum) :
    url = 'http://www.m-park.co.kr/buy/my_car_list.asp'
    data ={'sSearch': 1,
           'sListOrder': 'DemoDay_D',
           'sColorCode': value,
           'sPageSize': 40,
           'gotoPage': pagenum}
    res = requests.post(url,data=data)
    bs_obj = BeautifulSoup(res.text,'html.parser')
    
    return bs_obj
```

```python
def crawl_color(bs_obj, key):
    ul = bs_obj.find("ul", {"class":"theme_view on"})
    prd_info = ul.findAll("div",{"class":"car-info-s"})
    year= []
    year_1 = []
    km = []
    fuel = []
    # 옵션 추출
    for i in range(0,len(prd_info)) : 
        year.append(prd_info[i].text.split('\n',maxsplit = 4)[1])
        km.append(prd_info[i].text.split('\n',maxsplit = 4)[2])
        fuel.append(prd_info[i].text.split('\n',maxsplit = 4)[3])
    year_1 = [year[i].replace('년식',"") for i in range(0,len(year))]
    
    car_manu = []
    car_model_1 = []
    car_model_2 = []
    car_price = []
    
    prd_name = ul.findAll("div",{"class":"title-box"})
    # 모델명 추출
    car_model = [prd_name[i].find("h1").text for i in range(0,len(prd_name))]
    car_model_2 = [prd_name[i].find("h2").text for i in range(0,len(prd_name))]
    for i in range(0, len(car_model)) :
        car_manu.append(car_model[i].split("\xa0", maxsplit=1)[0])
        car_model_1.append(car_model[i].split("\xa0", maxsplit=1)[1])

    # 가격 추출
    prd_price = ul.findAll("div",{"class":"price-info"})
    car_price = [prd_price[i].find("span",{"class":"point"}).text for i in range(0,len(prd_price))]

    df = pd.DataFrame()
    df['제조사'] = car_manu
    df['모델명'] = car_model_1
    df['옵션명'] = car_model_2
    df['연식'] = year_1
    df['주행거리'] = km
    df['연료'] = fuel
    df['색상'] = [key for i in range(0,len(df))]
    df['가격'] = car_price
    return df
```

```python
bs_obj = bring_obj('006007', '1')
df = crawl_color(bs_obj, '검정')
df
```
