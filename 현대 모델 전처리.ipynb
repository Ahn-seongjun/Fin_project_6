## 전처리

```python
# 빈 열 생성
df['model']= ""
df['trim']= pd.Series() 
```

```python
# info의 데이터 빼내서 새 칼럼 만들기

for i in range(len(df['info'])):
    if "스타일" in df['info'][i]:
        df["trim"][i] = "1"
    elif "벨류 플러스" in df['info'][i]:
        df["trim"][i] = "1"
    elif "케어플러스" in df['info'][i]:
        df["trim"][i] = "1"
    elif "스마트" in df['info'][i]:
        df["trim"][i] = "1"
    elif "모던 베이직" in df['info'][i]:
        df["trim"][i] = "1"
    elif "모던" in df['info'][i]:
        df["trim"][i] = "1"
    elif "프리미엄" in df['info'][i]:
        df["trim"][i] = "2"
    elif "프리미엄스페셜" in df['info'][i]:
        df["trim"][i] = "2"
    elif "익스클루시브" in df['info'][i]:
        df["trim"][i] = "3"
    elif "익스클루시브 스페셜" in df['info'][i]:
        df["trim"][i] = "3"
    elif "셀러브리티" in df['info'][i]:
        df["trim"][i] = "3" 
    elif "프레스티지" in df['info'][i]:
        df["trim"][i] = "4" 
    elif "인스퍼레이션" in df['info'][i]:
        df["trim"][i] = "4" 
    elif "디럭스" in df['info'][i]:
        df["trim"][i] = "1" 
    elif "럭셔리" in df['info'][i]:
        df["trim"][i] = "2" 
    elif "프리미어" in df['info'][i]:
        df["trim"][i] = "3" 
    elif "프라임" in df['info'][i]:
        df["trim"][i] = "4" 
  
df.head()
```

## 작은 데이터 삭제

```python
df=df.groupby('type').filter(lambda x : len(x)>10)
df['type'].value_counts()
```

```python
df=df.groupby('name').filter(lambda x : len(x)>10)
df['name'].value_counts()
```

```python
df.drop(['info','model'], axis=1,inplace=True)
df.dropna(axis=0,inplace=True)
```

```python
df['type'] = df.apply(lambda x: x['type'].replace('(일반인 구입)',''), axis = 1)
df['type'] = df.apply(lambda x: x['type'].replace('가솔린+전기','하이브리드'), axis = 1)
df['type'] = df.apply(lambda x: x['type'].strip(), axis = 1)
```
