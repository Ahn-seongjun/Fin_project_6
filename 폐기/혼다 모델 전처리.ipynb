```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv("./data/혼다.csv")
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head(30)
```

```python
# 데이터 확인 용도로 사용
df['info'].value_counts().to_csv("혼다요약.csv")
```

```python
df['info']=df['info'].str.replace(pat='AWD', repl='')
df['info']=df['info'].str.replace(pat='2WD', repl='')
df['info']=df['info'].str.replace(pat='4WD', repl='')
df['info']=df['info'].str.replace(pat='하이브리드', repl='')
df['info']=df['info'].str.replace(pat='  ', repl='')
df['info']=df['info'].str.strip()
```

```python
df=df.groupby('info').filter(lambda x : len(x)>1)
```

```python
df['trim']= df['info']
df.drop(['info'], axis=1,inplace=True)
```

```python
df['type'] = df.apply(lambda x: x['type'].replace('(일반인 구입)',''), axis = 1)
df['type'] = df.apply(lambda x: x['type'].replace('가솔린+전기','하이브리드'), axis = 1)
df['type'] = df.apply(lambda x: x['type'].strip(), axis = 1)
```

```python
df.to_csv("혼다_문혁.csv")
```
